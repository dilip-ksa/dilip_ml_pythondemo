```python
##################  DAY 2 #########################
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
diabetes_ds = pd.read_csv('diabetes.csv')
```

```python
diabetes_ds.head()
```

```python
diabetes_ds.info()
```

```python
diabetes_ds.shape
```

```python
diabetes_ds.describe()
```

```python
diabetes_ds['Outcome'].value_counts()
```

```python
#Separating features and the target(Outcome)

features = diabetes_ds.drop(columns='Outcome',axis=1)
target = diabetes_ds['Outcome']
```

```python
features
```

```python
target.head()
```

```python
# Scaler calculation
scaler = StandardScaler()
scaler.fit(features)
```

```python
standardized_data= scaler.transform(features)
standardized_data
```

```python
features_data = standardized_data
features_data
```

```python
# Split the data using train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_data, target, test_size=0.2, random_state= 2)
```

```python
print(features_data.shape, x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
# Train the model
classifier = LogisticRegression(random_state=0, max_iter=1000)
classifier.fit(x_train, y_train)
```

```python
# Predict target for x_test dataset
x_test_prediction = classifier.predict(x_test)
```

```python
# Test accuracy of x_test_prediction with y_test
test_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy score of the test data : ', test_accuracy)
```

```python
input_data = (5,166,72,19,175,25.8,0.58,51)
input_as_array = np.asarray(input_data) # changing tuple into array
input_data_reshaped = input_as_array.reshape(1, -1) # reshape determines no of column and sets it

std_data = scaler.transform(input_data_reshaped)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
    print('The person is not diabetic')
else:
    print('The person is diabetic')
```
