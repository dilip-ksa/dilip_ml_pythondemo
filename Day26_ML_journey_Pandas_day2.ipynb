```python
student = {'Name': ['Ram', 'Radha', 'Mohan'], 'Age': [20,21,21], 'Marks': [89,90,88]}
df = pd.DataFrame(student)
print(type(df))
df
```

```python
ls = [10,20,30,40]
df = pd.DataFrame(ls)
print(type(df))
df
```

```python
df1 = pd.DataFrame(ls, columns=['Marks'])
print(type(df1))
df1
```

```python
#Custom row name
df = pd.DataFrame(student, index=['Row1','Row2','Row3'])
print(type(df))
df
```

```python
#Filter on column
df = pd.DataFrame(student, index=['Row1','Row2','Row3'], columns=['Name', 'Marks'])
print(type(df))
df
```

```python
#With unknown column
df = pd.DataFrame(student, index=['Row1','Row2','Row3'], columns=['Name', 'Marks', 'C1'])
print(type(df))
df
```

```python
#datatype
df = pd.DataFrame(student, index=['Row1','Row2','Row3'], dtype='str')
print(type(df))
df
```

```python
print(df.dtypes)
print(df.columns)
print(df.size)
print(df.shape)
print(df.ndim)
print(df.values)
```
